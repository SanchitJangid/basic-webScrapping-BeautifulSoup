#### Let's firslty Do all this With 1st Book,and then put it in a loop for all books

```python



import requests
from bs4 import BeautifulSoup


# Passing 1st page of books website
re = requests.get('https://books.toscrape.com/')

xc = BeautifulSoup(re.text,'html.parser')
print(xc.prettify())
```

```python
# Data related to books is present in this file , so we are just printing it firstly
data.find(class_ = 'product_pod' ) 
```

```python
# From above we can see that the Book URL is present in <h3><a>,Href = url , so now

xc.h3.a
```

```python
# To get Only URL , 

print(xc.h3.a['href'])
```

```python

# 1st Book's URL
base_url = 'https://books.toscrape.com/'

b1_url = base_url + xc.h3.a['href']
print(b1_url)
```

```python
# 1] To get the title

# get the book's data

rer = requests.get("https://books.toscrape.com/catalogue/a-light-in-the-attic_1000/index.html")

bdata = BeautifulSoup(rer.text,'html.parser')
print(bdata.prettify())
```

```python
# 1] To get the title

bdata.h1.string

#therefore
title = bdata.h1.string
title
```

```python
# 2] To get the price

price = bdata.find(class_ = 'price_color').string
price
```

```python
# 3] Quantity

quantity = data.find(class_ = 'instock availability')

#Therefore
qty = quantity.contents[-1].strip()
print(qty)
# idk why it's not showing , in stock value, it's showing in the website , 
# but it's not showing here
```

```python
 # , Now print all , 
print(title)
print(b1_url)
print(price)
print(qty)
```

```python
# NOW
print(title)
print(b1_url)
print(price)
print(qty)
```

#### NOW , let's organize our Data

```python
booksDetails = []
booksDetails.append([title,b1_url,price,qty])
booksDetails
```

```python
# SO it's a 2x2 list , such that each row will contain info of a book , and adn books will go
# Column by column
```

#### To save data as CSV , using Data Frame

```python
import pandas as pd

df = pd.DataFrame(booksDetails,columns = ['title','link','price','Quantity in stocks'])
df
```
